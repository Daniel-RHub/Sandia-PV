# Pandas File Input / Output

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
```

```python
url = 'http://apmonitor.com/che263/uploads/Main/tclab.txt'
x = pd.read_csv(url)
```

```python
print(x.head(10))
print(x.tail(10))
```

```python
s = x.describe()
print(s)
s.to_excel('Second Test.xlsx')
```

### Plot Data

```python
# Short cut -- When writing something often
tm = x['Time (sec)']

# 2 rows, 1 column, first subplot
plt.figure()
figure(figsize = (15,10))
plt.subplot(2,1,1)
plt.title('TCLab');
plt.plot(tm, x['Temperature 1'], label = r'$T_1$');
plt.plot(tm, x['Temperature 2'], label = r'$T_2$');
plt.legend(loc = 'best');
plt.ylabel('Temp (C)');

# Black and Red line 'k-' 'r--'
plt.subplot(2,1,2)
plt.plot(tm, x['Heater 1'], 'r--', label = r'$H_1$');
plt.plot(tm, x['Heater 2'], 'k-', label = r'$H_2$');
plt.ylabel('Heater (%)');
plt.xlabel('Time (sec)');
plt.legend(loc = 'best');

# Save as a PNG
# DPI (dots per inch)
plt.savefig('myPlot.png', dpi = 600, transparent = True)
```
